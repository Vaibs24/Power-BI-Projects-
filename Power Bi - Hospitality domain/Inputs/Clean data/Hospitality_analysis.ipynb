Problem Statement - Provide required insights for marketing generation team from past data as per the required data, representation and format style provided on mockup document.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
room = pd.read_csv('/content/sample_data/Hospltality data/dim_rooms.csv')
hotels = pd.read_csv('/content/sample_data/Hospltality data/dim_hotels.csv')
date = pd.read_csv('/content/sample_data/Hospltality data/dim_date.csv')
agg_book = pd.read_csv('/content/sample_data/Hospltality data/fact_aggregated_bookings.csv')
book = pd.read_csv('/content/sample_data/Hospltality data/fact_bookings.csv')
```

```python
book.info()
```

```python
book.isnull().sum()
```

```python
book['booking_date'] = pd.to_datetime(book['booking_date'], format= '%Y-%m-%d')
book['check_in_date'] = pd.to_datetime(book['check_in_date'], format= '%Y-%m-%d')
book['checkout_date'] = pd.to_datetime(book['checkout_date'], format= '%Y-%m-%d')
```

```python
book_new = book.merge(room, how= 'left', left_on= 'room_category', right_on= 'room_id')
del book_new['room_id']
book_new
```

```python
date.isnull().sum()
```

```python
book_new.to_csv('Book_final.csv')
```

```python
date_update = date.copy()
date_update[['day','month','year']] = date_update['date'].str.split('-', expand= True)
date_update
```

```python
date_update['month'].unique()
date_update['month'] = date_update['month'].replace(['May', 'Jun', 'Jul'],[5,6,7])
date_update
```

```python
date_update['year'].unique()
date_update['year'].replace(['22'], [2022], inplace= True)
```

```python
date_update['day'].astype(int)
date_update['date_format'] = pd.to_datetime(date_update[['year','month','day']], format= '%Y%m%d')
date_update
```

```python
date_update.duplicated().sum()
date_update.isnull().sum()
```

```python
date_update.to_csv('date_final.csv')
```

```python
agg_book.info()
agg_book_update = agg_book.copy()
agg_book_update[['day','month','year']] = agg_book_update['check_in_date'].str.split('-', expand= True)
agg_book_update
```

```python
agg_book_update['year'].replace(['22'], [2022], inplace= True)
agg_book_update['month'].replace(['May', 'Jun', 'Jul'],[5,6,7], inplace= True)
agg_book_update['day'].astype(int)
```

```python
agg_book_update['check_in_date_format'] = pd.to_datetime(agg_book_update[['year','month','day']], format= '%Y%m%d')
agg_book_update
```

```python
del agg_book_update['day']
```

```python
agg_book_update.drop(['month','year'], axis= 1, inplace= True)
agg_book_update
```

```python
agg_book_update.to_csv('agg_book_update.csv')
```
